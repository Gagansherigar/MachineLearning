```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
credit_card_data=pd.read_csv('/content/creditcard.csv')
```

```python
credit_card_data.isnull().sum()
```

```python
#distribution of legit and fraud
credit_card_data['Class'].value_counts()
```

data is not balanced ,make them balanced


```python
legit=credit_card_data[credit_card_data.Class==0]
fraud=credit_card_data[credit_card_data.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
fraud.Amount.describe(

)
```

```python
credit_card_data.groupby('Class').mean()
```

Undersampling

```python
legit_sample=legit.sample(n=173)
```

join legit_sample and fraud to make a good data

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head(

)
```

```python
new_dataset.shape
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

Logistic regression

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

Evaluation

```python
X_train_predictgion=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_predictgion,Y_train)
```

```python
print("the accuracy of training data is",training_data_accuracy)
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("the accuracy of test data is",test_data_accuracy)
```
